```python
# SVD
import numpy as np 
import pandas as pd
from sklearn.preprocessing import StandardScaler
from numpy import cov
from numpy.linalg import eig
from numpy.linalg import multi_dot
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("C:\\Users\\Ipsita\\Downloads\\train.csv")
```

```python
X = train
```

```python
#X = np.array([[4,0],[3,-5]])
```

```python
#X
X.head()
```

```python
#transposeA=A.dot(A.T)
transX=X.T
```

```python
transX
#transX.head()
```

```python
# X * X trans
x=X.dot(transX)
eig_val_x,eig_vec_x=np.linalg.eigh(x)
```

```python
eig_val_x
```

```python
eig_vec_x
```

```python
# X trans * X
y=X.T.dot(X)
eig_val_y,eig_vec_y= np.linalg.eigh(y)
```

```python
eig_val_y
```

```python
eig_vec_y
```

```python
idx=eig_val_y.argsort()[::-1]
eig_val_y=eig_val_y[idx]
eig_vec_y=eig_vec_y[idx]  
```

```python
eig_val_y
```

```python
V=np.matrix(eig_vec_y)
```

```python
V
```

```python
#Singular values
Sin_val=np.sqrt(eig_val_y)
```

```python
Sin_val
```

```python
a = np.array(Sin_val)
d = np.diag(a) # diagonal matrix of singular values
```

```python
S=np.matrix(d)
```

```python
S
```

```python
S_inv=np.linalg.inv(S)
```

```python
S_inv
```

```python
U=  X.dot(V).dot(S_inv)
```

```python
U
```

```python
# Reconstructed Matrix
M=U.dot(S).dot(V.T) 
```

```python
M
```

```python
error=np.subtract(M,X)
```

```python
error
```
